```python
import pathlib
import json

import utm
import numpy as np
import pandas as pd
```

Setup directory. May need to update based on current working directory location compared to rawdata.



```python
data_name = "TestData"
raw_data_directory = pathlib.Path.cwd() / "RawData" / data_name

raw_data_directory.exists()
```

Collect information about sensor location.
One col for the sensor ID, the column name for the dataset, then X column then Y column.
If X and Y is Long and Lat, enable the conversion.
Remember: Lat is Y, Long is X (i.e usually opposite) 


```python
station_info = pd.read_csv(raw_data_directory / "xycoords.csv")
station_info
```

```python
sensor_id_colname = "Sensor"
sensor_x_name = "X"
sensor_y_name = "Y"
is_latlong = False

sensor_locations = station_info[[sensor_id_colname, sensor_x_name, sensor_y_name]].values
sensor_location_map = {str(r[0]): ((r[1], r[2]) if not is_latlong else utm.from_latlon(r[2], r[1])) for r in sensor_locations}
sensor_location_map
```

Collect readings and merge aux data if applicable

```python
readings = pd.read_csv(raw_data_directory / "TestDataReadings.csv")
readings
```

```python
readings.columns
```

sensor IDs should be their column names

```python
sensor_IDs = list(sensor_location_map.keys())
drop_sensor_IDs = []
for s in drop_sensor_IDs:
    readings = readings.drop(s, axis = 1)
```

```python
readings = readings.dropna()
readings
```

```python
date_col = "date_time"
readings[date_col] = pd.to_datetime(readings[date_col])
readings
```

Get Aux data, skip if already included in readings and just set full = readings

```python
aux = pd.read_csv(raw_data_directory / "TestDataAux.csv")
aux
```

```python
aux_drop_columns = []
for c in aux_drop_columns:
    aux = aux.drop(c, axis = 1)
aux
```

```python
aux_date_time_col = "date_time"
aux[aux_date_time_col] = pd.to_datetime(aux[aux_date_time_col])
aux
```

Merge and interpolate

```python
full = pd.merge(readings, aux, how = 'left', left_on = date_col, right_on = aux_date_time_col)
full = full.set_index(date_col)
full = full.interpolate(method = "time")
full = full.fillna(method = "bfill")
full
```

```python
# If aux already in readings
# full = readings
full = full.dropna()
full.shape
```

```python
full.to_csv(f"{raw_data_directory / data_name}_full.csv")
print(sensor_IDs)
print({k:sensor_location_map[k] for k in sensor_location_map})
with open(f"{raw_data_directory / data_name}_sensors.json", 'w') as f:
    json.dump({k:sensor_location_map[k] for k in sensor_location_map if k in sensor_IDs}, f)
```
